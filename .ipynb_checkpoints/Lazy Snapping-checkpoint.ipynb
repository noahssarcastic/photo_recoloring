```python
import cv2
import numpy as np
from numpy.random import choice
import maxflow
from matplotlib.backend_bases import MouseButton
```

```python
def show_image(img):
    """
    Display image consistent with interface.
    
    :param ndarray img: image to display
    """
    
    import matplotlib.pyplot as plt
    %matplotlib inline
    fig = plt.figure(figsize=(10,10))
    plt.axis('off')
    plt.imshow(img, cmap='gray')
    plt.show()
    
    
def show_selection(img, foreground_lines, background_lines):
    """
    Display image with selections.
    
    :param ndarray img: image to display
    :param list foreground_lines: list of lists, stores each line as [[x1, x2...], [y1, y2...]]
    :param list background_lines: list of lists, stores each line as [[x1, x2...], [y1, y2...]]
    """
    
    import matplotlib.pyplot as plt
    %matplotlib inline
    fig = plt.figure(figsize=(10,10))
    plt.axis('off')
    plt.imshow(img, cmap='gray')
    
    for line in foreground_lines:
        [xs, ys] = line
        plt.plot(xs, ys, '-r', linewidth=5)
        
    for line in background_lines:
        [xs, ys] = line
        plt.plot(xs, ys, '-b', linewidth=5)
        
    plt.show()
```

```python
def _get_selection(img, foreground=True, foreground_lines=[]):
    """
    Make selections on an image.
    
    :param ndarray img: image to make selections on
    :param bool foreground: is a foreground selection
    :param list foreground_lines: list of lists, stores each line as [[x1, x2...], [y1, y2...]]
    :returns: list of lists, stores each line as [[x1, x2...], [y1, y2...]]
    :rtype: list
    """
    
    import matplotlib.pyplot as plt
    %matplotlib notebook
    
    fig = plt.figure(figsize=(10,10))
    plt.axis('off')
    plt.imshow(img, cmap='gray')

    if foreground:
        fig.set_label('Select Foreground')
        color = 'r'
    else: # background selection
        fig.set_label('Select Background')
        color = 'b'
        for line in foreground_lines:
            [xs, ys] = line
            plt.plot(xs, ys, '-r', linewidth=5)
            
    plt.show()

    lines = []

    def on_mouse_down(event):
        x = event.xdata
        y = event.ydata
        
        plt.plot(x, y, '+' + color)
        
        lines.append([[],[]])
        lines[-1][0].append(x)
        lines[-1][1].append(y)
        
    def on_mouse_move(event):
        if event.button==MouseButton.LEFT:
            x = event.xdata
            y = event.ydata
            lines[-1][0].append(x)
            lines[-1][1].append(y)
            plt.plot(x, y, '+' + color)
        
    def on_mouse_up(event):
        xs = lines[-1][0]
        ys = lines[-1][1]
        plt.plot(xs, ys, '-' + color, linewidth=5)

    fig.canvas.mpl_connect('button_press_event', on_mouse_down)
    fig.canvas.mpl_connect('button_release_event', on_mouse_up)
    fig.canvas.mpl_connect('motion_notify_event', on_mouse_move)
    
    return lines


def get_foreground(img):
    """
    Make foreground selections on an image.
    
    :param ndarray img: image to make selections on
    :returns: list of lists, stores each line as [[x1, x2...], [y1, y2...]]
    :rtype: list
    """
    
    return _get_selection(img, foreground=True, foreground_lines=[])
    
    
def get_background(img, foreground_lines):
    """
    Make foreground selections on an image.
    
    :param ndarray img: image to make selections on
    :param list foreground_lines: list of lists, stores each line as [[x1, x2...], [y1, y2...]]
    :returns: list of lists, stores each line as [[x1, x2...], [y1, y2...]]
    :rtype: list
    """
    
    return _get_selection(img, foreground=False, foreground_lines=foreground_lines)
```

```python
img = cv2.cvtColor(cv2.imread('samples/toy_problem.png'), cv2.COLOR_BGR2GRAY).astype('double') / 255.0
```

```python
foreground_lines = get_foreground(img)
```

```python
background_lines = get_background(img, foreground_lines)
```

```python
def get_seeds(img, lines):
    """
    Get the pixels that are intersected by selection points.
    
    :param ndarray img: domain of pixels
    :param list lines: list of lists, stores each line as [[x1, x2...], [y1, y2...]]
    :returns: coordinates of seed pixels in image space
    :rtype: (list, list)
    """
    
    rows = []
    cols = []
    for line in lines:
        [xs, ys] = line
        for i in range(len(xs)):
            x = xs[i]
            y = ys[i]
            rows.append(int(round(y)))
            cols.append(int(round(x)))
    return rows, cols
          
    
def get_mask(rows, cols, seed_rows, seed_cols):
    """
    Given image dimensions and list of seeds return a mask occluding other pixels.
    
    :param int rows: number of rows in image space
    :param int cols: number of columns in image space
    :param list seed_rows: list of rows of each seed pixel
    :param list seed_rows: list of columns of each seed pixel
    :returns: mask occluding non seed pixels
    :rtype: ndarray
    """
    
    mask = np.zeros((rows,cols))
    for i in range(len(seed_rows)):
        r = seed_rows[i]
        c = seed_cols[i]
        mask[r,c] = 1
    return mask
```

```python
# calculate F and B
f_seed_rows, f_seed_cols = get_seeds(img, foreground_lines)
b_seed_rows, b_seed_cols = get_seeds(img, background_lines)
rows, cols = img.shape
f_mask = get_mask(rows, cols, f_seed_rows, f_seed_cols)
b_mask = get_mask(rows, cols, b_seed_rows, b_seed_cols)

# display seed pixels for sanity check
full_mask = np.clip(f_mask+b_mask, 0, 1)
show_image(img*full_mask)
```

```python
def _find_cluster(point, centroids):
    """
    Find the closest centroid.
    
    :param point: data point
    :param ndarray centroids: 1d numpy array containing cluster centroids
    :returns: index of the cluster the point belongs to
    :rtype: int
    """
    
    dt = centroids.dtype
    diffs = np.array([abs(c - point) for c in centroids], dtype=dt)
    cluster_index = np.argmin(diffs)
    return cluster_index


def _get_cluster_means(points, centroids):
    """
    Calculate k means from centroids and points.

    :param ndarray points: 1d numpy array containing data points to be averaged
    :param ndarray centroids: 1d numpy array containing cluster centroids
    :returns: k means of each cluster
    :rtype: ndarray
    """
    
    k = len(centroids)
    dt = points.dtype
    
    # get mean of each cluster
    sums = np.zeros(k, dtype=dt)
    counts = np.zeros(k, dtype=np.int64)
    for p in points:
        cluster_index = _find_cluster(p, centroids)
        sums[cluster_index] += p
        counts[cluster_index] += 1
    
    k_means = sums / counts
    return k_means


def _get_k_means_variance(points, k_means):
    """
    Calculate the sum of variances of the k clusters.
    
    :param ndarray points: 1d numpy array containing data points in all clusters
    :param ndarray k_means: 1d numpy array containing cluster averages
    :returns: sum of variances of clusters
    """
    
    k = len(k_means)
    dt = points.dtype
    
    # get sum of square differences for each cluster
    ssds = np.zeros(k, dtype=dt)
    counts = np.zeros(k, dtype=np.int64)
    for p in points:
        cluster_index = _find_cluster(p, k_means)
        mean = k_means[cluster_index]
        ssds[cluster_index] += (p - mean)**2
        counts[cluster_index] += 1
        
    variances = ssds / counts
    return variances.sum()
    
    
    
def k_means_1D(points, k=64, repetitions=5, tol=10e-3):
    """
    Iteratively calculate k means and return the best.

    :param ndarray points: 1d numpy array containing data points to be averaged
    :param int k: number of centroids
    :param int repetitions: number of iterations of algorithm 
    :param float tol: tolerance used to detect convergence 
    :returns: k means with the lowest sum of variances
    :rtype: ndarray
    """
    
    n = len(points)
    dt = points.dtype
    
    best_k_means = np.zeros(k, dtype=dt)
    best_variance = np.inf
    
    # repeat and save k_means with lowest sum of variances
    for r in range(repetitions):
        centroids = np.zeros(k, dtype=dt)
        k_means = np.zeros(k, dtype=dt)
        
        seeds = np.random.choice(points, k, replace=False)
        k_means = seeds
        
        # optimize until change is less than tol
        while(np.absolute(centroids - k_means).max() > tol):
            centroids = k_means
            k_means = _get_cluster_means(points, centroids)
        
        variance = _get_k_means_variance(points, k_means)
        
        # save best k_means
        if variance < best_variance:
            best_k_means = k_means
            best_variance = variance
            
    return best_k_means  
```
