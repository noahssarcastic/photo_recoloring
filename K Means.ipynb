```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def _find_cluster(point, centroids):
    """
    Find the closest centroid.
    
    :param point: data point
    :param ndarray centroids: 1d numpy array containing cluster centroids
    :returns: index of the cluster the point belongs to
    :rtype: int
    """
    
    dt = centroids.dtype
    diffs = np.array([abs(c - point) for c in centroids], dtype=dt)
    cluster_index = np.argmin(diffs)
    return cluster_index


def _get_cluster_means(points, centroids):
    """
    Calculate k means from centroids and points.

    :param ndarray points: 1d numpy array containing data points to be averaged
    :param ndarray centroids: 1d numpy array containing cluster centroids
    :returns: k means of each cluster
    :rtype: ndarray
    """
    
    k = len(centroids)
    dt = points.dtype
    
    sums = np.zeros(k, dtype=dt)
    counts = np.zeros(k, dtype=np.int64)
    for p in points:
        cluster_index = _find_cluster(p, centroids)
        sums[cluster_index] += p
        counts[cluster_index] += 1
    
    k_means = sums / counts
    return k_means


def _get_k_means_variance(points, k_means):
    """
    Calculate the sum of variances of the k clusters.
    
    :param ndarray points: 1d numpy array containing data points in all clusters
    :param ndarray k_means: 1d numpy array containing cluster averages
    :returns: sum of variances of clusters
    """
    
    k = len(k_means)
    dt = points.dtype
    
    ssds = np.zeros(k, dtype=dt)
    counts = np.zeros(k, dtype=np.int64)
    for p in points:
        cluster_index = _find_cluster(p, k_means)
        mean = k_means[cluster_index]
        ssds[cluster_index] += (p - mean)**2
        counts[cluster_index] += 1
        
    variances = ssds / counts
    return variances.sum()
    
    
    
def k_means_1D(points, k=64, repetitions=5, tol=10e-3):
    """
    Iteratively calculate k means and return the best.

    :param ndarray points: 1d numpy array containing data points to be averaged
    :param int k: number of centroids
    :param int repetitions: number of iterations of algorithm 
    :param float tol: tolerance used to detect convergence 
    :returns: k means with the lowest sum of variances
    :rtype: ndarray
    """
    
    n = len(points)
    dt = points.dtype
    
    best_k_means = np.zeros(k, dtype=dt)
    best_variance = np.inf
    
    for r in range(repetitions):
        centroids = np.zeros(k, dtype=dt)
        k_means = np.zeros(k, dtype=dt)
        
        seeds = np.random.choice(points, k, replace=False)
        k_means = seeds
        
        plot_colors = "bgrcmyk"
        marker_size = 20
        plt.figure(figsize=(5,5))
        plt.plot(points, np.zeros(n), "k.", markersize=marker_size)
        plt.plot(k_means, np.zeros(k), "k|", markersize=marker_size)
        plot_height = 1
        
        while(np.absolute(centroids - k_means).max() > tol):
            centroids = k_means
            k_means = _get_cluster_means(points, centroids)
            
            for i, km in enumerate(k_means):
                plt.plot(km, plot_height, plot_colors[i] + "|", markersize=marker_size)
            for p in points:
                cluster_index = _find_cluster(p, k_means)
                plt.plot(p, plot_height, plot_colors[cluster_index] + ".", markersize=marker_size)
            plot_height += 1
        
        plt.show()
        
        variance = _get_k_means_variance(points, k_means)
        
        if variance < best_variance:
            best_k_means = k_means
            best_variance = variance
            
    return best_k_means
```

```python
data = np.random.rand(50)
k_means_1D(data, k=5, repetitions=5)
```
